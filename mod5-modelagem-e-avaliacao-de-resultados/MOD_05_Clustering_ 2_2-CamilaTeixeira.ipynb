## Import dos pacotes

```python
import pandas as pd

import seaborn as sns 
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

from sklearn.cluster import DBSCAN

from sklearn.metrics import silhouette_score

import gc 
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
dados = pd.read_csv('new_train.csv', sep=',')
```

```python
dados.tail()
```

```python
explicativas = dados.drop(columns=['y'], axis=1).copy()
```

```python
explicativas.dtypes
```

```python
explicativas.tail()
```

```python
expl_num = explicativas[['age','duration','campaign','pdays','previous']]
```

```python
expl_num.isnull().sum()
```

## Transformação por padronização

```python
scale = StandardScaler()
```

```python
expl_num_scale = scale.fit_transform(expl_num)
```

```python
expl_num_scale
```

## O algoritmo

```python
db = DBSCAN(eps=0.2)
```

```python
db
```

```python
explicativas['DB_eps_02'] = db.fit_predict(expl_num_scale)
```

```python
explicativas['DB_eps_02'].nunique()
```

Foram criados 110 grupos diferentes com o eps=2.

## Avaliação de melhor eps por meio do silhouette_score

```python
eps = [3,5]

for ep in eps:
  model = DBSCAN(eps=ep)
  predicao = model.fit_predict(expl_num_scale)

  score = silhouette_score(expl_num_scale, predicao)

  print("O valor da silhouette_score é {} para um número de eps igual a {}".format(score, ep))
```

```python
gc.collect()
```

```python
db = DBSCAN(eps=3)
```

```python
explicativas['DB_eps_3'] = db.fit_predict(expl_num_scale)
```

```python
explicativas['DB_eps_3'].value_counts()
```
